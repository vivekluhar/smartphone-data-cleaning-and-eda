## 1. Summary Of Data
The dataset contains 1,020 smartphone records with 11 features, including model name, price, rating, SIM specifications, processor details, RAM and storage information, battery capacity, display specifications, camera configuration, memory card support, and operating system. Most variables are stored as text because they combine multiple pieces of information within a single field (for example, RAM with storage and battery with charging details). The dataset represents smartphones from various brands and covers a wide range of hardware and software characteristics.

Overall, the dataset is largely complete and suitable for exploratory analysis of smartphone specifications and market trends. However, a few attributes contain missing values, particularly the rating column, along with a small number of missing entries in camera, memory card, and operating system fields. The data includes both numerical and categorical information embedded in text fields, providing detailed product descriptions that can support analyses related to pricing, performance, storage, battery features, display quality, and operating system distribution across smartphones.

## 2. Columns Description
#### **Table** -> `smartphone`:
1. `model` :- Contains the smartphone's brand and model name. This column uniquely identifies each smartphone in the dataset.

2. `price` :- Represents the selling price of the smartphone, generally expressed in Indian Rupees (₹). It can be used to analyze pricing trends across different brands and specifications.

3. `rating` :- Shows the overall user or expert rating assigned to the smartphone. Higher ratings generally indicate better customer satisfaction or product performance.

4. `sim` :- Provides information about SIM card support, such as Dual SIM capability, 4G, 5G, VoLTE, and other network-related features available on the device.

5. `processor` :- Contains details about the chipset or processor powering the smartphone. It may include processor name, manufacturer, core count, and clock speed information.

6. `ram_battery` :- Stores both RAM and battery-related specifications in a single field. Typically includes RAM size (GB), battery capacity (mAh), and charging technology details.

7. `display` :- Describes the smartphone screen characteristics, including display size, resolution, refresh rate, panel type, and other visual features.

8. `camera` :- Contains information about the rear and front camera setup. This may include megapixel counts, multiple camera configurations, and camera-related features.

9. `card` :- Indicates whether external memory expansion is supported and may contain details about microSD card compatibility and maximum expandable storage.

10. `os` :- Represents the operating system running on the smartphone, such as Android or other mobile operating systems, along with version information when available.
